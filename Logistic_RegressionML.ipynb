```python
# Supervised ML - Classification --- Logistic Regression
# Sigmoid Function
import numpy as np
from sklearn.linear_model import LogisticRegression 
#dataset
X = np.array([[1],[2],[3],[4],[5],[6]]) 
y = np.array([0,0,0,1,1,1])
#train model
model = LogisticRegression()
model.fit(X,y)
```

```python
# predict_proba() - predict probabilities 
y_prob = model.predict_proba(X)[:,1] #[:,1] all rows from column 1 
y_prob
```

```python
a = model.coef_[0][0]
b = model.intercept_[0]

print(f'a (slope) :  {a}')
print(f'b (intercept) :  {b}')
# a = 1.12 ---> model suggests 'hours studied' highly correlated to chance of passing
# if a = 0.2 effect/correlation weaker

# b = -3.9 
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
```

```python
data = pd.read_csv('diabetes.csv')
data.head() 
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
x = data.drop('Outcome', axis = 1) 
y = data['Outcome']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000) 
model.fit(x_train,y_train) 
```

```python
# predict_proba() stands for predict probabilities 
# predict_proba() returns 2 columns for the negative and positive class 
# For classification, get predicted probabilities for positive labels only 
y_prob= model.predict_proba(x_test)[:,1]
y_prob
```

```python
y_prob = model.predict(x_train) 
y_prob
```


This code creates a comprehensive table showing predictions alongside actual values (if available) and can include relevant features from your test data. You can customize which columns to include based on your specific needs.

```python
# get predicted values 
y_pred = model.predict(x_test) 
y_pred
```

```python
ev1l = pd.DataFrame({'Probabilities':y_prob,'Predicted_values':y_test})
ev1l.head(50)
```

```python
from sklearn.metrics import confusion_matrix , classification_report 
cm = confusion_matrix(y_test,y_pred) 
cm 

#                             Actual_Positive        Actual_Negative 
#    Predicted Positive           78   TP               21  FP
#    Predicted_Negative           18   FN               37  TN
#                                                          Total Number of Observations: 154  115 correct 39 Wrong
```

```python
y_pred.shape
```

```python
print('Classification Report\n') 
print(classification_report(y_test,y_pred)) 
```

```python
sns.heatmap(cm,annot = True,fmt ='d',cmap = 'Blues') # fmt ---> d - display decimal/int value
plt.xlabel('Predicted') 
plt.ylabel('Actual')
plt.title ('Predicted vs Actual') 
plt.show()
```
